```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
```

```python
dataset = pd.read_csv('C://Users//User//Desktop//Datasets for Machine Learning//Mall_Customers.csv')
X = dataset.iloc[:, [3,4]].values  # we could have taken the last 4 features but we took only last 2 as we want to visualize
# the dataser in the 2 Dimesional Space 
#print(X)
```

```python
plt.scatter(X[:,0] , X[:,1])
```

```python
#Now we will try to plot the Elbow Method in order to check the correct value where the Plot will break
from sklearn.cluster import KMeans
wcss = []
for i in range(1,11):
    kmeans = KMeans(n_clusters = i , init = 'k-means++', random_state = 42) # Here we try to fit model for each cluster 
    #n_clusters imploes the number of clusters and init = k-means++ is for coing up with the random initialization error 
    kmeans.fit(X) # Then we fit the model according to the above model
    wcss.append(kmeans.inertia_) # this inertia method basically calculates the value of the wcss value that we add to the
    #initially declared list wcss
plt.plot(range(1,11), wcss) # we use the range(1,11) as the x variable to depict the number of clusters and wcss as the 
# the calculated sum of square of distance beween the point and its corresponding Centroid
plt.title('The Elbow Method')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS  Value')
    
```

```python
kmeans = KMeans(n_clusters = 5 , init = 'k-means++', random_state = 42)
y_kmeans = kmeans.fit_predict(X) #This is a special method to directly predict the the values in the K mean clustering method

print(y_kmeans)
```

```python
#Visualising the K-mean Clusters 
plt.scatter(X[y_kmeans==0,0] , X[y_kmeans==0,1] , s = 100 , c='red' , label = 'cluster1' ) 
#Now we are basically doing is that we need to specify the x and y co-ordinates in the scatter function like we specify the 
#X[y_kmeans==0,0] means that we want all the clusters with the kmeans as 0 and only the fist column of it i.e the annual income
#column and same X[y_kmeans==0,1] specifies the columns Spending score to that x value
plt.scatter(X[y_kmeans==1,0] , X[y_kmeans==1,1] , s = 100 , c='black' , label = 'cluster2' )
plt.scatter(X[y_kmeans==2,0] , X[y_kmeans==2,1] , s = 100 , c='green' , label = 'cluster3' )
plt.scatter(X[y_kmeans==3,0] , X[y_kmeans==3,1] , s = 100 , c='cyan' , label = 'cluster4' )
plt.scatter(X[y_kmeans==4,0] , X[y_kmeans==4,1] , s = 100 , c='blue' , label = 'cluster5' )
plt.scatter(kmeans.cluster_centers_[:,0] , kmeans.cluster_centers_[:,1] , s= 200 , c= 'yellow' , label = 'centroids')
plt.legend()
#same as above we have the .cluster_centers_ containing the centers of the clusters we just need to specify their values of 
# x and y i.e kmeans.cluster_centers_[:,0]  corresponds to the x value or the first column of the centroid 
```
