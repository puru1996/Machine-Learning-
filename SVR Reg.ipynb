```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
```

```python
dataset = pd.read_csv('C://Users//User//Desktop//Position_Salaries.csv')
X= dataset.iloc[:,1:-1].values
y= dataset.iloc[:,-1].values
y=y.reshape(len(y),1)
print(X)
print (y)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
sc2 = StandardScaler()
X = sc.fit_transform(X)
y = sc2.fit_transform(y) 
print(X)
print(y)
```

```python
from sklearn.svm import SVR  
regressor = SVR(kernel = 'rbf')
regressor.fit(X,y)
```

```python
sc2.inverse_transform(regressor.predict(sc.transform(([[6.5]]))))
```

```python
plt.scatter(sc.inverse_transform(X) , sc2.inverse_transform(y))
plt.plot(sc.inverse_transform(X),sc2.inverse_transform(regressor.predict(X))) 
plt.xlabel('Position')
```
