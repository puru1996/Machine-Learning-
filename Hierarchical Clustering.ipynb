```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import scipy.cluster.hierarchy as sch
```

```python
dataset = pd.read_csv('C://Users//User//Desktop//Datasets for Machine Learning//Mall_Customers.csv')
X = dataset.iloc[:, [3,4]].values  # we could have taken the last 4 features but we took only last 2 as we want to visualize
# the dataser in the 2 Dimesional Space 
#print(X)
```

```python
dendrogram = sch.dendrogram(sch.linkage(X,method ='ward')) #we need to give two parameter 1 is the matrix of features on 
# which we need to perform the clustering technique and second the method of clustering here we choose the method with minimum
#variance i.e ward method 
plt.title('Dendrogram')
plt.xlabel('Customer')
plt.ylabel('Eucledian Distances')
plt.show()
```

```python
#Now from the Dendrogram i have to select the largest line which is not intercepted by any of the Horizontal lines with each
#horizonatal line extended to the maximum distance and next we have to draw a horzontal line throught the midpoint of that line and check how many
#vertical lines it itersects and that is the number of clusters
```

```python
from sklearn.cluster import AgglomerativeClustering
hc = AgglomerativeClustering(n_clusters = 5, affinity = 'euclidean', linkage = 'ward')
y_pred = hc.fit_predict(X)
```
